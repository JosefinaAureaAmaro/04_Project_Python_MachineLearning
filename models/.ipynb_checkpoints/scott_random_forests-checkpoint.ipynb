```python
#import dependencies 

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# read in CSV

raw_economic_2015_data = pd.read_csv('Machine_Learning_Dataset_2015.csv')
raw_economic_2015_data.head(40)

```

```python
dropcol_2015 = raw_economic_2015_data.dropna(axis='columns')
dropcol_2015
```

```python
list(dropcol_2015.columns) 
```

```python
narrow_2015_data = dropcol_2015.drop(["COU",
                                      "Country",
                                      "Literacy rate, adult total (% of people ages 15 and above)",
                                      "GINI index (World Bank estimate)_x",
                                      "Year",
                                      "Negative affect",
                                      "Positive affect",
                                      "Delivery Quality",
                                      "Standard deviation of ladder by country-year",
                                      "Standard deviation/Mean of ladder by country-year"], 
                                      axis=1)

narrow_2015_data
```

```python
# random trees can't take floats, so I'm changing all data types to integer

rev_2015_data = narrow_2015_data.astype('int64')
```

```python
# assign X (data) and y (target)

X = rev_2015_data.drop("Life Ladder", axis=1)
y = rev_2015_data["Life Ladder"]
print(X.shape, y.shape)
```

```python
# establish variable to hold feature names (the names of the remaining columns)

feature_names = X.columns
```

```python
# split data into training and testing 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# create a random forest classifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# calculate feature importance

importances = rf.feature_importances_
importances
```

```python
# sort the features by their importance

sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
