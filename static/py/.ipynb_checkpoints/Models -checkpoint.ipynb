```python
#Importing our libraries 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
```

```python
wdi = pd.read_csv('../resources/Machine_Learning_Dataset_2011.csv')
wdi.head()
```

```python
#removing columns that are similar to one another to remove multicollinearity from our dataset when we run our models.  

columns = ['GINI index (World Bank estimate), average 2000-16',
       'gini of household income reported in Gallup, by wp5-year',
       'Most people can be trusted, Gallup',
       'Most people can be trusted, WVS round 1981-1984',
       'Most people can be trusted, WVS round 1989-1993',
       'Most people can be trusted, WVS round 1994-1998',
       'Most people can be trusted, WVS round 1999-2004',
       'Most people can be trusted, WVS round 2005-2009',
       'Most people can be trusted, WVS round 2010-2014']

wdi = wdi.drop(columns, axis=1)

wdi.head()
```

```python
#This is an approximation which can add variance to the data set. 
#But it is better than having loss of the data. 
#Because our dataset is so small using this method may yeild better results compared to removal of rows and columns

wdi['GINI index (World Bank estimate)_y'] =  wdi['GINI index (World Bank estimate)_y'].replace(
np.NaN, wdi['GINI index (World Bank estimate)_y'].mean())
```

```python
wdi.isnull().sum()
```

Model 1: Multiple Linear Regression

Our data looks at multiple indicators that may affect the happiness of a population in a given country. We can test a multiple linear model to determine how the value of our "happiness" indicator is determined from multiple independent variables. 

As observed in our dataset, we already have a predicted value, "Life Ladder", or colloquially - a happiness index. A model that would predict a quantity would not be that useful for the data we have. Therefore, we can assume that regression models will not be as accurate.  

Our aim is to see how our features predict a label. This is where we can see the fundamental difference between classification and regression problems:

Classification is about predicting a label and regression is about predicting a quantity.

```python
X = wdi[["Avg. Wages (Annual)", 
         "Avg. Work Hours (Annual)",
         "Literacy rate, adult total (% of people ages 15 and above)",
         "Probability of dying at age 5-14 years (per 1,000 children age 5)",
         "Freedom to make life choices"
         ]]

y = wdi["Life Ladder"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

We now fit our model to the linear regression model by training the model with our independent variables and dependent variable.

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to predict a value
predicted = model.predict(X)

# Score the prediction with mse and r2
mse = mean_squared_error(y, predicted)
r2 = r2_score(y, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```

Model 2: Random Forest 

When decision trees are used, they use our features(columns), to split our observations into groups that are different from one another. The members in these groups, however, are similar to one another. 

The observations made by the decision tree then allow it to make conclusions about the target. In our case, the Life Ladder metric. 

Random Forest, takes many decision trees and uses the observations made by them to make a prediction about a class. The class that appears the most often is the predicted class. 

```python
#Looking at summary statistics to look for anomalies. 

wdi.describe
```

Running a Random Forest Classifier

```python
wdi = wdi.drop(["Country","COU","Year"], axis=1)
```

```python
# changing the format of our data types of the values in our dataframe so that we can use it for random forest classifier. 
# random forest classifier does not take float 64 

wdi = wdi.astype('int64')
```

```python
y = wdi["Life Ladder"]
```

```python
X = wdi.drop(["Life Ladder"], axis=1)
feature_names = X.columns
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

As expected, our classification model performed much better than the linear regreession model. 

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
feature_importance = rf.feature_importances_

# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.figure(figsize=(20, 30))
plt.subplot(1, 2, 2)

plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, feature_names[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.savefig('rf_classifier_features.png')
plt.show()
```

Building a Random Forest Regressor. Because our model has floats this model could be a better fit. 
However, since our aim is to predict a class, we can expect this model to underperform compared to the classifier. 

```python
X = wdi.drop(["Life Ladder"], axis=1)
```

```python
y= wdi["Life Ladder"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train, y_train)
regressor.score(X_test, y_test)
```

The regression model does not score as well as the classifier. 

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
importances = list(regressor.feature_importances_)

# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_names, importances)]

# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```

```python
x_values = list(range(len(importances)))

# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')

# Tick labels for x axis
plt.xticks(x_values, feature_names, rotation='vertical')

# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');

```

Model 3: Support Vector Machines 

```python
wdi.head()
```

```python
wdi_svm = wdi.astype('int64')
```

```python
X = wdi_svm.drop(["Life Ladder"], axis=1)
feature_names = X.columns
```

```python
y= wdi_svm["Life Ladder"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#when running the svm with a linear kernel, the code ran indefinately. I suspect this happened because our data does not fit a
# linear model. Therefore, I chose a polynomial kernel to allow the SVM to learn as a non-linear model. 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
model = SVC(kernel='poly', degree=8)
model.fit(X_train, y_train)
```

```python
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, predictions)

sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=feature_names,
            yticklabels=feature_names)

plt.xlabel('true label')
plt.ylabel('predicted label');
```
